```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# prep data
nb_datapoints = 100

np.random.seed(42)
df = pd.DataFrame(np.random.randint(2, size=(nb_datapoints,2)), columns=['dimension 1', 'dimension 2'])
np.random.seed(42)
df['val'] = np.random.rand(nb_datapoints)
```

```python
to_groupby = ['dimension 1', 'dimension 2']
y_col = 'val'


fig, ax = plt.subplots() 

df.boxplot(by=to_groupby, column=[y_col], ax=ax)

# adding scattered data points to the plot
counter = 1
for i, g in df.groupby(by=to_groupby):
    y = g[y_col]
    x = np.random.normal(counter, 0.04, size=len(y))
    ax.plot(x, y, 'g.', alpha=0.2)    
    
    counter += 1

# ax.set_ylim([0.50,1.0])
plt.title('predictions for subgroups of data')
# plt.ylabel('')
plt.xlabel('groups')
fig.set_size_inches(10, 8) 

# plt.savefig('plots/fig.svg')

plt.show()

plt.gcf().clear()
mpl.pyplot.close("all")
```
